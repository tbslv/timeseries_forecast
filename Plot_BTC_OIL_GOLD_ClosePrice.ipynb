```python
data = pd.read_csv(
    '/Users/tobiasleva/Desktop/Local_repo/archive/data_151223.csv',
    index_col = 0 )
```

```python
fig,axs = plt.subplots(4,1,figsize=(5,7))
sns.lineplot(data=data,x='Date',y='Close_btc',ax=axs[0])

axs[0].set_xlabel('')
axs[0].set_xticks([])
axs[0].set_xticklabels([])
axs[0].set_ylabel('Bitcoin Close Price ($)')
axs[0].set_ylim(0,)
sns.lineplot(data=data,x='Date',y='Close_gold',ax=axs[1])
axs[1].set_xlabel('')
axs[1].set_xticks([])
axs[1].set_xticklabels([])
axs[1].set_ylabel('Gold Close Price ($)')
axs[1].set_ylim(1000,)
sns.lineplot(data=data,x='Date',y='Close_oil',ax=axs[2])
axs[2].set_xlabel('')
axs[2].set_xticks([])
axs[2].set_xticklabels([])
axs[2].set_ylabel('Crude Oil Close Price ($)')
axs[2].set_ylim(-50,)
sns.lineplot(data=data,x='Date',y='Value',ax=axs[3])

axs[3].set_xticks(['2018-01-01','2019-01-01','2020-01-01','2021-01-01','2022-01-01',
                  '2023-01-01'])
axs[3].set_xticklabels(['2018','2019','2020','2021','2022','2023'])
axs[3].set_ylim(-6,)
axs[3].set_xlabel('Year')
axs[3].set_ylabel('Sentiment Score')

plt.tight_layout()
```

```python
plt.style.use('dark_background')

# Creating subplots
fig, axs = plt.subplots(4, 1, figsize=(5, 7))

# Plotting Bitcoin Close Price
sns.lineplot(data=data, x='Date', y='Close_btc', ax=axs[0], color='red',label='Bitcoin',lw=1)
axs[0].legend(loc='upper left', frameon=False)
axs[0].set_xlabel('')
axs[0].set_xticks([])
axs[0].set_ylabel('Close Price ($)', fontsize=12)
axs[0].set_ylim(0,)

axs[0].set_xticks(['2018-01-01','2019-01-01','2020-01-01','2021-01-01','2022-01-01','2023-01-01'])
axs[0].set_xticklabels([])

# Plotting Gold Close Price
sns.lineplot(data=data, x='Date', y='Close_gold', ax=axs[1], color='gold',label='Gold',lw=1)
axs[1].legend(loc='upper left', frameon=False)
axs[1].set_xlabel('')
axs[1].set_xticks([])
axs[1].set_ylabel('Close Price ($)', fontsize=12)
axs[1].set_ylim(1000,)

axs[1].set_xticks(['2018-01-01','2019-01-01','2020-01-01','2021-01-01','2022-01-01','2023-01-01'])
axs[1].set_xticklabels([])

# Plotting Crude Oil Close Price
sns.lineplot(data=data, x='Date', y='Close_oil', ax=axs[2], color='white',label='Oil',lw=1)
axs[2].set_xlabel('')
axs[2].set_xticks([])
axs[2].set_ylabel('Close Price ($)', fontsize=12)
axs[2].set_ylim(-50,)
axs[2].axhline(0,ls='--',lw=.5,color='grey')
axs[2].legend(loc='upper left', frameon=False)

axs[2].set_xticks(['2018-01-01','2019-01-01','2020-01-01','2021-01-01','2022-01-01','2023-01-01'])
axs[2].set_xticklabels([])

# Plotting Sentiment Score
sns.lineplot(data=data, x='Date', y='Value', ax=axs[3], color='green',label='Sentiment',lw=1)
axs[3].set_xticks(['2018-01-01','2019-01-01','2020-01-01','2021-01-01','2022-01-01','2023-01-01'])
axs[3].set_xticklabels(['2018','2019','2020','2021','2022','2023'],fontsize=12)
axs[3].set_ylim(-6,)
axs[3].set_xlabel('Year',fontsize=12)
axs[3].set_ylabel('Sentiment Score', fontsize=12)
axs[3].axhline(0,ls='--',lw=.5,color='grey')
axs[3].legend(loc='upper left', frameon=False)
sns.despine()

plt.tight_layout()
plt.savefig(r'/Users/tobiasleva/Desktop/close_prices.png')

plt.show()
```
